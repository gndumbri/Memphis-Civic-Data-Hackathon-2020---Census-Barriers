# Method

Use Census Data > Determine Barriers > Highlight Barrier Distribution Geographically > Bust Barriers

```python
census['cluster'] = census['cluster'].astype(str)
fig = px.scatter_mapbox(census, lat="lat", lon="lon", hover_name="TractFIPS", hover_data=['Interview Ratio','Persuade Ratio','Contact Ratio','Locate Ratio'],
                        color='cluster', zoom=10, height=800)
fig.update_layout(mapbox_style="carto-darkmatter")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.update_layout(
    legend=dict(
        x=0,
        y=1,
        traceorder="normal",
        font=dict(
            family="sans-serif",
            size=20,
            color="black"
        ),
        bgcolor="White",
        bordercolor="Black",
        borderwidth=2
    )
)
fig.show()
```
